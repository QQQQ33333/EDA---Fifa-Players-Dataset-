2.Plot a bar chart of 5 top countries with the most number of players.

```python
df1 =pd.DataFrame({'Nationality':['England','Germany ','Spain','Argentina','France'],'Total players':[1662,1198,1072,937,914]})
```

```python
df1
```

```python
sns.barplot(x='Nationality',y='Total players',data=df1,color='g')
plt.title('TOP 5 NATIONS WITH MOST NO. OF PLAYERS',color='purple')
plt.xlabel('NATIONS',color='b')
plt.ylabel("Total no. of players",color="b")
plt.grid(axis='y', linestyle='--')
plt.show()
```

5.Who is the tallest player in the fifa?

```python
ht=fifa[['Name','Height']]
ht
```

```python
SORTht=ht.sort_values(by='Height',ascending =False)
T=SORTht.head()
T
```

```python
sns.scatterplot(x='Name',y='Height',data=T,color='olive',marker="o")
plt.title('PLAYERS WITH GREATEST HIGHT',color='purple')
plt.xlabel('NAME',color='b')
plt.ylabel("HEIGHT",color="b")
plt.show()
```

T. Holý	and	D. Hodzic are heighest among FIFA players.Both T. Holý	and	D. Hodzic	have 6'9 feet height.

 7.Which foot is most preferred by the players?Draw a bar chart for preferred foot

```python
foot=fifa['Preferred Foot'].value_counts()
foot
```

```python
sns.barplot(data=foot,color='yellow')
plt.title('PREFERRED FOOT BY PLAYERS',color='purple')
plt.show()
```

From above visualization, We can conclude that most of the players prefer their right foot to kick the ball. Near 14K(13948) players uses their right foot.

Left foot is preferred by only 4211 players.

3.Which player has the highest salary?

```python
fifa['Wage'] = fifa['Wage'].str.replace('€', '').str.replace('K', '000')
```

```python
fifa['Wage'] = fifa['Wage'].astype(float)
```

```python
fifa['Wage']
```

```python
fifa['Name'][fifa['Wage']==fifa['Wage'].max()]
```

Without any doubt, we can clearly say that L. Messi has highest salary among FIFA players.

4.Plot a histogram to get the salary range of the players.

```python
plt.hist('Wage',data=fifa,color='teal')
plt.title('Salary Range of the FIFA Player',color='m')
plt.xlabel('Salary')
plt.ylabel('Number of Players')
plt.xticks(rotation=90)
plt.show()
```

Most of the players are getting the salary less than or around 50000. Very few of them are getting salary greater than 50000.
